```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
digits=load_digits()
```

```python
print(digits)
```

```python
df=pd.DataFrame(digits.data)
df['Target']=digits.target
print(df.head())
```

```python
Kmeans=KMeans(n_clusters=10,random_state=42)
Kmeans.fit(digits.data)
```

```python
df['cluster']=Kmeans.labels_
print(df.head())
```

```python
pca=PCA(n_components=2)
pca_result=pca.fit_transform(digits.data)
```

```python
plt.scatter(pca_result[:,0],pca_result[:,1],c=df['cluster'],cmap='gray')
plt.xlabel('PCA component 1')
plt.ylabel('PCA component 2')
plt.title('PCA K-Means clustering on Digits Dataset')
plt.show()
```

```python
cluster_centers=Kmeans.cluster_centers_.reshape(10,8,8)
fig,axis=plt.subplots(2,5,figsize=(8,4))
for i,ax in enumerate(axis.flat):
  ax.imshow(cluster_centers[i],cmap='gray')
  ax.set_title(f'Cluster {i}')
  ax.axis('off')
plt.show()
```
